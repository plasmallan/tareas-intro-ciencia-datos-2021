```python
file=r'oec.csv'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(file)
```

```python
data.head()
```

```python
data.tail()
```

```python
groupedby=data.groupby('DiscoveryMethod').agg({'PlanetaryMassJpt':['mean','median','std','min','max','sum','count']})
```

```python
groupedby
```

```python
plt.bar(groupedby.index,groupedby.PlanetaryMassJpt['sum'])
```
